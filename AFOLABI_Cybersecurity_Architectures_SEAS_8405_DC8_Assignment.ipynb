**Equifax Data Breach (2017) - Base Architecture (On-premises)**

**SolarWinds Backdoor Attack (2020) - Base Architecture (On-prem DevOps)**
**Vulnerabilities Recap:**
1. Compromised Orion build environment with malicious updates

2. Lack of code integrity verification

3. No policy enforcement or tamper detection

```python
from diagrams import Diagram, Cluster
from diagrams.onprem.client import Users
from diagrams.onprem.network import Internet
from diagrams.onprem.vcs import Github
from diagrams.onprem.ci import Jenkins
from diagrams.onprem.compute import Server
from diagrams.onprem.security import Vault
from diagrams.onprem.monitoring import Prometheus
from diagrams.custom import Custom
from IPython.display import Image

# Use a placeholder icon for malware-infected build server
icon_url_malware = "https://raw.githubusercontent.com/mingrammer/diagrams/master/resources/onprem/security/security.png"

diagram_file = "solarwinds_architecture"

with Diagram("SolarWinds Base Architecture", filename=diagram_file, outformat="png", show=False):
    devs = Users("Developers")
    internet = Internet("Internet")

    with Cluster("DevOps Pipeline"):
        repo = Github("Source Code Repo")
        ci = Jenkins("Build Server")
        infected = Custom("Infected Artifact", icon_url_malware)
        vault = Vault("Secrets Vault")
        monitor = Prometheus("Monitoring")

        devs >> repo
        repo >> ci >> infected
        ci >> vault
        infected >> monitor

    internet >> infected

Image(filename=f"{diagram_file}.png")
```

**Capital One Data Breach (2019) - Base Architecture (AWS Cloud)**
**Vulnerabilities Recap: **
1. SSRF exploit on misconfigured WAF allowed metadata access

2. Over-privileged IAM roles led to S3 data exfiltration

3. Insufficient monitoring

```python
from diagrams import Diagram, Cluster
from diagrams.aws.compute import EC2
from diagrams.aws.database import RDS
from diagrams.aws.network import ELB, Route53
from diagrams.aws.security import WAF, IAM, Shield
from diagrams.onprem.network import Internet
from diagrams.custom import Custom
from IPython.display import Image

# Placeholder icon for SSRF exploit
icon_url_ssrf = "https://raw.githubusercontent.com/mingrammer/diagrams/master/resources/onprem/security/security.png"

diagram_file = "capitalone_architecture"

with Diagram("Capital One Base Architecture", filename=diagram_file, outformat="png", show=False):
    internet = Internet("Internet")
    waf = WAF("AWS WAF")
    lb = ELB("Elastic Load Balancer")

    with Cluster("AWS VPC"):
        ec2 = EC2("Web App EC2")
        rds = RDS("Customer PII (RDS)")
        ssrf = Custom("SSRF Exploit", icon_url_ssrf)
        iam = IAM("IAM Role Access")

    internet >> waf >> lb >> ec2
    ec2 >> ssrf >> iam >> rds

Image(filename=f"{diagram_file}.png")
```
